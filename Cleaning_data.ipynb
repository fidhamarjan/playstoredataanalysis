```python
import pandas as pd
import numpy as np
def convert_size(size):
    if size=='Varies with device':
        return float(0)
    elif 'M' in size:
        return float(size.replace('M', '')) * 1024
    elif 'k' in size:
        return float(size.replace('k',''))
    else:
        return np.nan
    
df['Size_KB'] = df['Size'].apply(convert_size)
```
