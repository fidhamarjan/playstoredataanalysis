QUESTIONS AND ANSWERS

1.Does a higher rating lead to more installs?

```python
plt.scatter(df['Installs'],df['Rating'])
plt.xlabel("Number of Installs")
plt.ylabel("Rating")
plt.title("Relation between Installs and Rating")
```

Higher rating does not always mean more installs,
More installs does not always mean better rating 

2.Are free apps more more popular than paid apps?

```python
plt.figure()
sns.barplot(x="Type",y="Installs",data=df)
plt.title("barplot")
```

Free apps get significantly more installs than paid apps

3.Which app categories have the highest number of installs?

```python
Category_Installs=df.groupby("Category")["Installs"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=Category_Installs.values,y=Category_Installs.index)
plt.xlabel("Number of Installs")
plt.ylabel=("Category")
plt.title("Total installs by category")
```

Game,Communication and Social are the highest installs categories

4.How does price affect the number of downloads?

```python
plt.figure()
sns.barplot(x="Type",y="Installs",data=df)
plt.title("barplot")
```

Free apps get significantly more installs than paid apps

5.Is there a relationship between the number of reviews and rating?

```python
sns.scatterplot(x= 'Reviews',y = 'Rating', data = df)
plt.xlabel("Number of Reviews")
plt.ylabel("Ratings")
plt.title("Relation between Reviews and Rating")
plt.show()
```

There is no relationship between number of reviews and rating
